```python
import os
import shutil
import numpy as np
import tensorflow as tf
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.applications import InceptionV3
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Dense, GlobalAveragePooling2D
from tensorflow.keras.optimizers import Adam,AdamW,Adamax,SGD
import matplotlib.pyplot as plt
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay, roc_auc_score, roc_curve, auc
import seaborn as sns
```

```python
base_dir = '../final_dataset1'
train_dir = os.path.join(base_dir, 'train')
val_dir = os.path.join(base_dir, 'val')
test_dir = os.path.join(base_dir, 'test')
```

```python
#Normalization
train_datagen = ImageDataGenerator(rescale=1./255)
val_datagen = ImageDataGenerator(rescale=1./255)
test_datagen = ImageDataGenerator(rescale=1./255)

#Train,Val and Test sets
train_generator = train_datagen.flow_from_directory(
    train_dir,
    target_size=(224, 224),
    batch_size=32,
    class_mode='binary',
)

val_generator = val_datagen.flow_from_directory(
    val_dir,
    target_size=(224, 224),
    batch_size=32,
    class_mode='binary',
)

test_generator = test_datagen.flow_from_directory(
    test_dir,
    target_size=(224, 224),
    batch_size=32,
    class_mode='binary',
    shuffle=False,
)
```

```python
base_model = InceptionV3(weights='imagenet', include_top=False, input_shape=(229, 229, 3))
```

```python
#Final layers set
x = base_model.output
x = GlobalAveragePooling2D()(x)
x = Dense(1024, activation='relu')(x)
output_layer = Dense(1, activation='sigmoid')(x)
```

```python
model = Model(inputs=base_model.input, outputs=output_layer)
```

```python
#Freeze early layers
for layer in base_model.layers:
    layer.trainable = False
model.compile(optimizer=Adam(learning_rate=0.001), loss='binary_crossentropy', metrics=['accuracy'])
```

```python
"""GRID SEARCH (optimizer + learning rate)
from tensorflow.keras.models import clone_model

lrs = [0.01, 0.001, 0.0001, 0.00001]
optimizers = ['Adam', 'AdamW', 'Adamax','SGD']

initial_model = clone_model(model)  
initial_model.set_weights(model.get_weights()) 

for opt in optimizers:
    for lr in lrs:
        print(f"Training for: {opt}, with learning rate: {lr}")
        
        model = clone_model(initial_model)  
        model.set_weights(initial_model.get_weights()) 
        
        if opt == 'Adam':
            optimizer = Adam(learning_rate=lr)
        elif opt == 'AdamW':
            optimizer = AdamW(learning_rate=lr)
        elif opt == 'Adamax':
            optimizer = Adamax(learning_rate=lr)
        elif opt == 'SGD':
            optimizer = SGD(learning_rate=lr)

        model.compile(optimizer=optimizer, loss='binary_crossentropy', metrics=['accuracy'])
        model.fit(train_generator, epochs=10, validation_data=val_generator)
        
        loss, acc = model.evaluate(test_generator)
        print(f"Optimizer: {opt}, Learning Rate: {lr}, Accuracy: {acc}, Loss: {loss}")
"""
```

```python
history = model.fit(
    train_generator,
    epochs=5,
    validation_data=val_generator
)
```

```python
test_loss, test_acc = model.evaluate(test_generator)
print(f'Test accuracy: {test_acc}')
```

```python
#Accuracy,Val accuracy, loss and val loss graphs
acc = history.history['accuracy']
val_acc =  history.history['val_accuracy']

loss = history.history['loss']
val_loss = history.history['val_loss']

epochs_range = range(5)

plt.figure(figsize=(10,5))
plt.subplot(1,2,1)
plt.plot(epochs_range,acc,label='Accuracy')
plt.plot(epochs_range,val_acc,label="Validation Accuracy")
plt.legend()

plt.subplot(1,2,2)
plt.plot(epochs_range,loss,label='Loss')
plt.plot(epochs_range,val_loss,label="Validation Loss")
plt.legend()
plt.show()
```

```python
predictions = model.predict(test_generator)
binary_predictions = (predictions > 0.5).astype(int).flatten()

true_labels = test_generator.classes
```

```python
#Confussion matrix
cm = confusion_matrix(true_labels, binary_predictions)

#Confusion matrix plot
plt.figure(figsize=(8, 6))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=['Breast Ultrasounds', 'Irrelevant Images'], yticklabels=['Breast Ultrasounds', 'Irrelevant Images'])
plt.xlabel('Predicted Label')
plt.ylabel('True Label')
plt.title('Confusion Matrix')
plt.show()
```
